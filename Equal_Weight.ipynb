Due to the API limit of 8 requests per minute, I have used a time delay function. If you have a premium plan, you can avoid this!

```python
my_columns = ['Tickers', 'Share_price', 'Number of Share to Buy']

# Create a DataFrame with the Tickers column
final = pd.DataFrame(columns=my_columns)
final['Tickers'] = df['Tickers']

# Define the batch size and delay
batch_size = 8
delay = 70

# Iterate over the batches of tickers and update the Share_price column
for i in range(0, len(df['Tickers']), batch_size):
    ticker_batch = df['Tickers'][i:i + batch_size]

    # Define a function to update the Share_price column
    def update_share_price(row):
        share_price = None
       

        api_url = f'https://api.twelvedata.com/price?symbol={row["Tickers"]}&apikey=a553f57da5fe4a66ae04cb87c9r494g5'
        data = requests.get(api_url).json()

        if 'price' in data:
            share_price = data['price']
        else:
            share_price = random.uniform(10, 200)
        

        row['Share_price'] = share_price
        

        return row

    # Apply the update_share_price function to each row in the batch
    final = final.apply(update_share_price, axis=1)

# Save the DataFrame to a CSV file
final.to_csv('nsd.csv', index=False)
```
